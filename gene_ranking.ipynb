```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gene_ranks_df = pd.read_csv('rankingSystem.csv')
gene_ranks_df
```

```python
gene_ranks_df['overall_score'] = (
    (gene_ranks_df['log_bc_ratio'] / max(gene_ranks_df['log_bc_ratio']) * 100) +
    (gene_ranks_df['Module Appearance Score'] / max(gene_ranks_df['Module Appearance Score']) * 100) +
    (gene_ranks_df['Cluster Y/N'] / max(gene_ranks_df['Cluster Y/N']) * 100) +
    (gene_ranks_df['Average Shortest Path Length'] / max(gene_ranks_df['Average Shortest Path Length']) * 100)
)
```

```python
ranked_genes = gene_ranks_df.sort_values('overall_score', ascending=False)
ranked_genes.to_csv('ranked_genes.csv')
```
